```python
labels = tf.constant(np.random.randint(0, 5, [10,10]), dtype=tf.int32)
```

```python
tf.reset_default_graph()
```

```python
sess=tf.Session()
    
print(sess.run(labels))
```

```python
l_uni = tf.unique_with_counts(tf.reshape(labels,[-1]))
weights = tf.reciprocal(tf.to_float(l_uni.count))
weights_normalized = weights * tf.to_float(tf.size(weights))/tf.reduce_sum(weights)
```

```python
sess.run(weights_normalized)
```

```python
weights_normalized.shape
```

```python
class_weights = tf.Variable([0.,0.,0.,0.,0.],dtype=tf.float32,trainable=False)
```

```python
class_weights = class_weights.assign(weights_normalized)
```

```python
class_weights
```

```python
sess.run(class_weights)
```
